```python
# Cell 1: Unzip Dataset with Nested Structure Handling
import zipfile
import os
import shutil

# Clear previous extraction if exists
if os.path.exists('training_set'):
    shutil.rmtree('training_set')
    print("ğŸ§¹ Cleared previous training_set directory")

zip_path = 'training_set.zip'
extract_to = 'training_set'

# Extract zip
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

# Handle nested structure
extracted_folders = os.listdir(extract_to)
print(f"ğŸ“‚ Extracted folders: {extracted_folders}")

# If there's only one folder inside, use it as the base path
if len(extracted_folders) == 1 and os.path.isdir(os.path.join(extract_to, extracted_folders[0])):
    base_path = os.path.join(extract_to, extracted_folders[0])
    print(f" Detected nested structure. Using base path: {base_path}")
else:
    base_path = extract_to
    print(" Using top-level extraction path")

# Cell 3: Define Paths with Nested Structure
benign_path = os.path.join(base_path, 'benign')
malignant_path = os.path.join(base_path, 'malignant')

print(f"\n Using benign path: {benign_path}")
print(f" Using malignant path: {malignant_path}")
print(f"Benign path exists: {os.path.exists(benign_path)}")
print(f"Malignant path exists: {os.path.exists(malignant_path)}")

# Find images with multiple patterns
image_patterns = ['*.[jJ][pP][gG]', '*.[pP][nN][gG]', '*.[jJ][pP][eE][gG]', '*.[bB][mM][pP]']
benign_images = []
malignant_images = []

for pattern in image_patterns:
    benign_images.extend(glob(os.path.join(benign_path, pattern)))
    malignant_images.extend(glob(os.path.join(malignant_path, pattern)))

# Recursive search if needed
if not benign_images:
    print("âš ï¸ No benign images found - trying recursive search")
    for pattern in image_patterns:
        benign_images.extend(glob(os.path.join(benign_path, '**', pattern), recursive=True))

if not malignant_images:
    print("âš ï¸ No malignant images found - trying recursive search")
    for pattern in image_patterns:
        malignant_images.extend(glob(os.path.join(malignant_path, '**', pattern), recursive=True))

print(f"\nâœ… Benign images found: {len(benign_images)}")
print(f"âœ… Malignant images found: {len(malignant_images)}")

if benign_images:
    print(f"Sample benign: {benign_images[0]}")
if malignant_images:
    print(f"Sample malignant: {malignant_images[0]}")

# Cell 4: Create DataFrame with Validation
if not benign_images and not malignant_images:
    # Show directory structure for debugging
    print("\nğŸ“‚ Directory structure:")
    for root, dirs, files in os.walk(base_path):
        level = root.replace(base_path, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f'{indent}{os.path.basename(root)}/')
        subindent = ' ' * 4 * (level + 1)
        for f in files[:3]:
            print(f'{subindent}{f}')
        if len(files) > 3:
            print(f'{subindent}...({len(files)} files)')

    raise RuntimeError("CRITICAL ERROR: No images found. Please check folder structure above")

# Rest of Cell 4 remains the same...

# Cell 4: Create DataFrame with Priority Labels
data = []

for path in benign_images:
    data.append((path, 'low'))  # benign = low priority

for path in malignant_images:
    priority = random.choice(['medium', 'high'])  # simulate priority
    data.append((path, priority))

image_df = pd.DataFrame(data, columns=['filepath', 'priority'])
label_map = {'low': 0, 'medium': 1, 'high': 2}
image_df['label'] = image_df['priority'].map(label_map)

print(image_df.head())

# Cell 5: Split Data
train_df, test_df = train_test_split(
    image_df, test_size=0.2, stratify=image_df['label'], random_state=42
)

print(f"Train size: {len(train_df)}, Test size: {len(test_df)}")

# Cell 6: Load and Preprocess Images
IMG_SIZE = (128, 128)

def load_and_preprocess(df):
    images = []
    labels = []
    for _, row in df.iterrows():
        try:
            img = load_img(row['filepath'], target_size=IMG_SIZE, color_mode='rgb')
            img_array = img_to_array(img) / 255.0
            images.append(img_array)
            labels.append(row['label'])
        except Exception as e:
            print(f"Error loading image {row['filepath']}: {e}")
    return np.array(images), to_categorical(np.array(labels), num_classes=3)

X_train, y_train = load_and_preprocess(train_df)
X_test, y_test = load_and_preprocess(test_df)

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

# Cell 7: Define CNN Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(3, activation='softmax')  # 3 priority levels
])

model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Cell 8: Train the Model
early_stop = EarlyStopping(patience=5, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop]
)

# Cell 9: Evaluate Model
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

print("Accuracy:", accuracy_score(y_true, y_pred))
print("F1 Score (macro):", f1_score(y_true, y_pred, average='macro'))
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=label_map.keys()))

# Cell 10: Plot Accuracy Curve
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
